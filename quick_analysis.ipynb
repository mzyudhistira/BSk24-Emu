```python
def analyse_varian_effect(varians, files):
    files.reverse()
    results = [run_analysis.analyse(file) for file in files]
    rms_dev = [result['rms_dev'] for result in results]
    std_diff = [result['std_diff'] for result in results]
    melted_mass_table = [result['melted_mass_table'] for result in results]

    output = {'rms_dev' : rms_dev, 'std_diff' : std_diff,
              'mmt':melted_mass_table} 

    return output
```

```python
df = aaa

# Group by 'id' column and split the dataframe
grouped = df.groupby('me_diff_over_std')

# Create a dictionary of dataframes
dfs = {id_: group for id_, group in grouped}

pkk = []

# Print the split dataframes
for id_, group in dfs.items():
    pke = {}
    pke['label'] = f'{id}*' + r'$\sigma$'
    pke['data'] = group
    pkk.append(pke)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mean = 0
std_dev = 1
num_samples = 1000

# Generate random data (normal distribution)
data = np.random.normal(mean, std_dev, num_samples)

# Create histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.6, color='b')

# Generate x values for the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)

# Plot normal distribution curve (PDF)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'k', linewidth=2)

# Add lines for mean and standard deviations
plt.axvline(mean, color='r', linestyle='dashed', linewidth=2)
plt.axvline(mean - std_dev, color='g', linestyle='dashed', linewidth=2)
plt.axvline(mean + std_dev, color='g', linestyle='dashed', linewidth=2)

# Labels and title
plt.xlabel(r'$M_\text{ML}$ (MeV)')
plt.ylabel('Frequency')

# Show plot
plt.show()
```
