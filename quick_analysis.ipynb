```python
def analyse_varian_effect(varians, files):
    files.reverse()
    results = [run_analysis.analyse(file) for file in files]
    rms_dev = [result["rms_dev"] for result in results]
    std_diff = [result["std_diff"] for result in results]
    melted_mass_table = [result["melted_mass_table"] for result in results]

    output = {"rms_dev": rms_dev, "std_diff": std_diff, "mmt": melted_mass_table}

    return output
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mean = 0
std_dev = 1
num_samples = 1000

# Generate random data (normal distribution)
data = np.random.normal(mean, std_dev, num_samples)

# Create histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.6, color="b")

# Generate x values for the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)

# Plot normal distribution curve (PDF)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, "k", linewidth=2)

# Add lines for mean and standard deviations
plt.axvline(mean, color="r", linestyle="dashed", linewidth=2)
plt.axvline(mean - std_dev, color="g", linestyle="dashed", linewidth=2)
plt.axvline(mean + std_dev, color="g", linestyle="dashed", linewidth=2)

# Labels and title
plt.xlabel(r"$M_\text{ML}$ (MeV)")
plt.ylabel("Frequency")

# Show plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Example DataFrame
df = pd.DataFrame(
    {"a": [1, 2, 3, 4, 5], "b": [2, 3, 4, 5, 6], "x": [1.5, 2.7, 3.2, 4.8, -1.3]}
)

# Apply floor function and clip values to the range [0, 3]
df["x"] = np.floor(df["x"]).clip(0, 3)

# Define custom colors for each value of x
color_map = {0: "blue", 1: "green", 2: "orange", 3: "red"}

# Create a new column for the colors based on 'x' values
df["color"] = df["x"].map(color_map)

# Create a scatter plot
plt.figure(figsize=(8, 6))

# Scatter plot using the manually assigned colors
scatter = plt.scatter(df["a"], df["b"], c=df["color"], s=100)

# Add custom legend for the color mapping
import matplotlib.lines as mlines

legend_labels = ["0 (Blue)", "1 (Green)", "2 (Orange)", "3 (Red)"]
legend_handles = [
    mlines.Line2D(
        [0], [0], marker="o", color="w", markerfacecolor="blue", markersize=10
    ),
    mlines.Line2D(
        [0], [0], marker="o", color="w", markerfacecolor="green", markersize=10
    ),
    mlines.Line2D(
        [0], [0], marker="o", color="w", markerfacecolor="orange", markersize=10
    ),
    mlines.Line2D(
        [0], [0], marker="o", color="w", markerfacecolor="red", markersize=10
    ),
]

plt.legend(legend_handles, legend_labels, title="x Values")

# Label the axes
plt.xlabel("a")
plt.ylabel("b")

# Show the plot
plt.title("Scatter plot with custom colors and legend based on x values")
plt.show()
```
